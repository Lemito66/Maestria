- 1.1 Despliega la frecuencia de aparición de cada palabra del texto. 

```python
def word_frecuency(data):
    frecuency = {}
    for word in data.split():
        frecuency[word] = frecuency.get(word, 0) + 1

    return frecuency 
```

```python
print(word_frecuency(data))
```

- 1.2 Cuál es la oración más larga? 

```python
def separate_dot(data):
    data = data.split('.')
    return data
```

```python
print(separate_dot(data))
```

```python
def long_sentence(data):
    data = separate_dot(data)
    long = 0
    for sentence in data:
        if len(sentence) > long:
            long = len(sentence)
            sentence_long = sentence
    return sentence_long
```

```python
print(long_sentence(data))
```

- 1.3 Cuántas oraciones empiezan con ‘A’?

```python
def how_many_words_start_with(data, letter):
    data = separate_dot(data)
    count = 0
    for sentence in data:
        for word in sentence.split():
            if word[0] == letter.upper():
                count += 1
    return count
```

```python
print(data)
```

```python
print(how_many_words_start_with(data, 'A'))
```

- 1.4 Por cada oración, ¿cuál es la palabra más larga?

```python
# Por cada oración que tengo, cual es la palabra más larga
def long_word_in_sentence(data):
    data = separate_dot(data)
    words = []
    for sentence in data:
        long = 0
        for word in sentence.split():
            if len(word) > long:
                long = len(word)
                word_long = word
        words.append(word_long)
                
    return words
```

- 2.1 ¿Cuántas veces aparece la palabra ‘coronavirus’?  

```python
def count_word_occurrences(data, word):
    count = 0
    for sentence in data:
        count += sentence.lower().count(word.lower())
    return count
```

```python
print(count_word_occurrences(data_corpus, 'coronavirus'))
```
