```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
import statsmodels.api as sm
```

```python
# read csv file
data = pd.read_csv("HousingData.csv")
data.columns
# convertir a minusculas
data.columns = [x.lower() for x in data.columns]
data.columns
```

```python
data_boston = load_data("Boston")
columns_boston = data_boston.columns
columns_boston

# tomas todo expepto la ultima
y = data_boston['medv']

intercept = pd.DataFrame({'intercept': np.ones(data_boston.shape[0])})

X = pd.concat([intercept, data_boston[columns_boston[:-1]]], axis=1)

X
```

```python
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```
