```python
# Escriba el código aquí
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# leer archivo de excel 
df = pd.read_excel('recienNacido.xlsx')
df.head()
```

```python
# Hacer un grafico normal donde la edad es mi x y los puntos es el normal y sobrepeso
plt.scatter(df['Edad'], df['Sobrepeso'], color='red')
plt.xlabel('Edad')
plt.ylabel('Sobrepeso')
plt.title('Edad vs Sobrepeso')
plt.show()
```

```python
# Crear el gráfico de dispersión
plt.scatter(df['Edad'], df['Sobrepeso'], color='red', label='Sobrepeso')
plt.scatter(df['Edad'], df['Normal'], color='blue', label='Normal')

# Trazar las líneas rectas para Sobrepeso y Normal
plt.plot(df['Edad'], df['Sobrepeso'], color='red', linestyle='-', linewidth=1)
plt.plot(df['Edad'], df['Normal'], color='blue', linestyle='-', linewidth=1)

# Etiquetas y título
plt.xlabel('Edad')
plt.ylabel('Sobrepeso / Normal')
plt.title('Edad vs Sobrepeso / Normal')
plt.legend()

plt.show()
```
