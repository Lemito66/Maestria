## **1. Variables aleatorias discretas**

### **1.1. Variable Bernoulli**

```python
# aplicando la formula
def pmf_bernoulli(probability):
    k_values = [0, 1]
    px_values = []
    for k in k_values:
        px = probability**k * (1 - probability)**(1 - k)
        px_values.append(px)
        
    fig = plt.figure(figsize=(3,3))
    ax = plt.axes()
    sns.barplot(x=k_values, y= px_values, color='b', ax=ax)
    plt.show()
```

```python
pmf_bernoulli(0.97)
```

### **1.2. Variable Binomial**

```python
# aplicando la formula
def pmf_binomial(n_events, probability):
    k_values = np.arange(0, n_events + 1)
    probability_values = factorial(n_events) / (factorial(k_values) * factorial(n_events - k_values)) * probability**k_values * (1 - probability)**(n_events - k_values)
    fig = plt.figure(figsize=(13,3))
    ax = plt.axes()
    sns.barplot(x=k_values, y= probability_values, color='b', ax=ax)
    plt.show()
```

```python
# saber la probilidad de que salga 3 veces cara en 10 lanzamientos
pmf_binomial(10, 0.5)
```

```python
# aplicando librerias python
pmf_binomial(30, 0.2)
```
