### **1.1. Variable uniforme**

```python
# aplicando librerias python
numbers = uniform.rvs(loc=0, scale=5,size=1000000, random_state=42)
sns.histplot(numbers, stat='probability')
```

### **2.1. Una variable sigue una distribucion normal con promedio 0 y desviación estandar de 1. Cual es la probabilidad de que la variable sea igual a 1. Cual es la probabilidad de que la variable tome valores mayores o igual a 0. Cual es la probabilidad de que la variable tome valores entre -1 y 1.**

```python
# probabilidad de que sea 1
norm.pdf(1, loc=0, scale=1)

```

```python
# probabilidad que tome valores mayores o iguales a 1
1 - norm.cdf(1, loc=0, scale=1)
```
