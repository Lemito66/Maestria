```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### **1. Estimar el valor de pi**

```python
n_simulations = 10000000
inside = []
outside = []
for i in range(n_simulations):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    distance = np.sqrt((x-0)** 2 + (y-0)** 2)
    if distance <= 1:
        inside.append((x, y))
    else:
        outside.append((x, y))

```

```python
plt.figure(figsize=(6, 6))
plt.scatter([i[0] for i in inside], [i[1] for i in inside], color='blue', s=0.1)
plt.scatter([i[0] for i in outside], [i[1] for i in outside], color='red', s=0.1)
plt.show()
```

```python
pi = 4 * len(inside) / (len(inside) + len(outside))
```

```python
n_simulations = 100000
inside = 0
predictions = np.zeros(n_simulations)
predictions =  np.zeros(n_simulations)

for i in range(n_simulations):
    x = np.random.uniform(-1, 1)
    y = np.random.uniform(-1, 1)
    distance = np.sqrt((x-0)** 2 + (y-0)** 2)
    if distance <= 1:
        inside = inside + 1
    predictions[i] = 4 * inside / (i + 1)

predictions
```

```python
plt.semilogx(predictions, color='blue')
plt.axhline(y=np.pi, color='red', linestyle='-')
plt.show()
```

```python
plt.loglog(np.arange(1, n_simulations + 1), np.abs(predictions - np.pi), color='blue')
plt.loglog(np.arange(1, n_simulations + 1), 1/np.sqrt(np.arange(1, n_simulations + 1)), color='red') # 1/sqrt(x) es la pendiente de la recta
plt.show()
```
