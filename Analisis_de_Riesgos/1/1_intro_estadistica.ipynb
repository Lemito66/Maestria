## **2. Generalizando distribuciones**

### **2.1. Variables discretas**


```python
n = 30
p = 0.5
numbers = binom.rvs(n, p, size=10000, random_state=42)
sns.histplot(x=numbers, discrete=True, stat="probability")
plt.show()
```

```python
sns.histplot(x=numbers, discrete=True, cumulative=True, stat="probability")
plt.show()
```

### **2.2. Variables continuas**


```python
x_values = np.linspace(-4, 4, 100)
pdf = norm.pdf(x_values, 0, 1)
sns.lineplot(x=x_values, y=pdf)
plt.ylabel("probability")
plt.show()
```

```python
cdf = norm.cdf(x_values, 0, 1)
sns.lineplot(x=x_values, y=cdf)
plt.ylabel("probability")
plt.show()
```

```python
numbers = norm.rvs(loc=0, size=10000, random_state=42)
sns.histplot(x=numbers, stat="probability")
plt.show()
```

```python
sns.histplot(x=numbers, stat="probability", cumulative=True)
plt.show()
```

```python
def plot_multiple_hist(dataframe, var_list, n_rows, n_cols, fig_size=(9, 10), discrete=False, stat='probability'):
  fig, ax = plt.subplots(n_rows, n_cols, figsize=fig_size)
  ax = np.ravel(ax)
  for i in range(len(var_list)):
    sns.histplot(x=dataframe[var_list[i]], discrete=discrete, stat=stat, ax=ax[i])
    if discrete:
      ax[i].set_xticks(dataframe[var_list[i]].unique())
    ax[i].set_title(var_list[i])
    ax[i].grid()
  ax = np.reshape(ax, (n_rows, n_cols))
  fig.subplots_adjust(hspace=0.6)
  plt.grid(None)
  plt.show()
```

```python
var_continuas = ['Price', "KM", "HP", "Age_08_04"]
plot_multiple_hist(df, var_continuas, 2,2, (9,5))
```

### **4.1. La junta directiva ve como un riesgo el gasto en publicidad y ofertas de la empresa. Como gerente de ventas, te solicitan optimizar las poblaciones objetivo de la publicidad y ofertas a fin de reducir gastos. Basado unicamente en las distribuciones de ganancias ('Revenue'), que propondras?**

```python
column_target = 'Revenue'
column_test = 'Customer Age'
df_group = df.groupby(column_test, as_index=False).agg({column_target: 'sum'})
df_group['porcentaje'] =df_group[column_target] / df_group[column_target].sum()
df_group['porcentaje_acumulado'] = df_group['porcentaje'].cumsum()
df_group.head()
```

```python
fig = plt.figure(figsize=(20, 6))
ax = plt.axes()
sns.barplot(x=column_test, y='porcentaje_acumulado', data=df_group, ax=ax)
```

```python
column_target = 'Revenue'
column_test = 'Sub Category'
df_group = df.groupby(column_test, as_index=False).agg({column_target: 'sum'}).sort_values(column_target, ascending=False)
df_group['porcentaje'] =df_group[column_target] / df_group[column_target].sum()
df_group['porcentaje_acumulado'] = df_group['porcentaje'].cumsum()
df_group.head()
```

```python
fig = plt.figure(figsize=(20, 6))
ax = plt.axes()
sns.barplot(x=column_test, y='porcentaje_acumulado', data=df_group, ax=ax)
```
