## **1. Al recibir un cargamento de microchips, 800 chips son seleccionados al azar. Si 3 o menos microchips resultan defectuosos, todo el cargamento es ingresado sin inspeccionar el resto del cargamento. Si 4 o más microchips resultan defectuosos, se inspecciona todos los microchips del cargamento y el costo que lo cubre el proveedor. Asuma que la probabilidad estimada de chips defectuosos es del 0.1%. Encontrar la probabilidad de que el cargamento de microchips sea aceptado. Resolver considerando una distribucion binomial y tambien la aproximación de Poisson.**

```python
n = 800
probability = 0.001

binom.cdf(3, n, probability)

print("Probabilidad de 3 o menos: {}".format(binom.cdf(3, n, probability)))
print("Probabilidad de 4 o más: {}".format(1 - binom.cdf(3, n, probability)))

# ahora con poisson
lambda_ = n * probability
poisson.cdf(3, lambda_)
print("Probabilidad de 3 o menos: {}".format(poisson.cdf(3, lambda_)))
print("Probabilidad de 4 o más: {}".format(1 - poisson.cdf(3, lambda_)))
```

## **2. Se sabe que aproximadamente 30 clientes por hora visitan una tienda. Encontrar:**

### **- El número esperado (expected value) de clientes que visiten la tienda en los primeros 10 minutos de una hora.**

### **- La probabilidad de que 4 clientes visiten la tienda en los primeros 10 minutos de una hora.**

### **- La probabilidad de que 35 o más clientes visiten la tienda en una hora.**



```python
# 30 clientes por hora visitan una tienda. El npumero esperado que visiten en los primeros 10 minutos de una hora

lambda_ = 30 * 10 / 60
print("Probabilidad de 10 clientes en 10 minutos: {}".format(poisson.pmf(10, lambda_)))

# probabilidad de que 4 clientes visiten la tienda en los prmeros 10 minutos de una hora 
print("Probabilidad de 4 clientes en 10 minutos: {}".format(poisson.pmf(4, lambda_)))

# probabilidad de que 35 o más visiten la tienda en una hora 
print("Probabilidad de 35 o más clientes en una hora: {}".format(1 - poisson.cdf(34, 30)))

# graficamos la distribución de poisson con lambda = 30

x = np.arange(0, 50)
y = poisson.pmf(x, lambda_)



sns.barplot(x=x, y=y, color='blue', alpha=0.7)
plt.xlabel('Número de clientes')
plt.ylabel('Probabilidad')
plt.title('Distribución de Poisson con $\lambda = 30$')
plt.show()




```
