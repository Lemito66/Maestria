## **2. Simular la evolución del conteo de problación. Para esto considerar que el limite mínimo de individuos es 0 y el límite máximo de individuos es 100. Las tasas de nacimiento y muerte se han estimado en 0.005 por individuo y por año. El estado inicial de la población es 25 individuos.**

```python
limit_min = 0
limit_max = 100
population_init = 25
```

```python
b_rate = 0.005 # naciomimento por individuo por año
d_rate = 0.005 # muerte por individuo por año

```

```python
n_times = 1000

population_count = [population_init]
population = population_init
for t in range(n_times -1):
    if limit_min < population < limit_max:
        birth = 1 if np.random.rand() <= b_rate * population else 0
        death = 1 if np.random.rand() <= d_rate * population else 0
        population += birth - death
        
    else:
        population = population
    population_count.append(population)

print(population_count)
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))
ax.plot(population_count)
ax.set_xlabel('Time')
ax.set_ylabel('Population')
ax.set_title('Population growth')
plt.show()
```

```python
n_simulations = 10000
population_sim = []

for n in range(n_simulations):
    n_times = 1000
    population = population_init
    population_count = [population_init]
    for t in range(n_times -1):
        if limit_min < population < limit_max:
            birth = 1 if np.random.rand() <= b_rate * population else 0
            death = 1 if np.random.rand() <= d_rate * population else 0
            population += birth - death

        else:
            population = population
        population_count.append(population)
    population_sim.append(population_count)

print(np.mean(population_sim, axis=0))
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,5))
ax.plot(population_count)
ax.set_xlabel('Time')
ax.set_ylabel('Population')
ax.set_title('Population growth')

for i in population_sim:
    ax.plot(i)
plt.show()
```
