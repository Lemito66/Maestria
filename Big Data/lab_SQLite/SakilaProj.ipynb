```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect('sakila_master.db')
```

```python
print("Connection Successful",connection)
```

```python
cur = connection.cursor() 
```

```python
# listando las tablas de la bdd
query = ''' SELECT name FROM sqlite_master WHERE type='table' '''
sth = cur.execute(query) 
results = sth.fetchall() 
print(results)
```

```python
query = ''' SELECT * FROM language ''' 
sth = cur.execute(query) 
results = sth.fetchall() 
print(results)
```

```python
#todos los clientes que han rentado al menos un film
pd.read_sql(""" SELECT rental_date, customer_id
                FROM rental 
                INNER JOIN inventory
                ON rental.inventory_id = inventory.inventory_id 
            """, 
            connection).head(35)
```

```python
# quÃ© estamos desplegando?
pd.read_sql(""" SELECT rental_date, rental.customer_id, inventory.film_id, 
                title, first_name, last_name, amount
                FROM rental
                INNER JOIN inventory
                ON rental.inventory_id = inventory.inventory_id
                INNER JOIN film
                ON film.film_id = inventory.film_id
                INNER JOIN payment
                ON rental.rental_id = payment.rental_id
                INNER JOIN customer
                ON rental.customer_id = customer.customer_id
            """, 
            connection).head(35)
```

```python
pd.read_sql(""" SELECT *
                FROM film
                WHERE title LIKE '%BLANKET%' """,
           connection, index_col=['film_id'])
```

```python
#todos los clientes que han rentado un film dado
pd.read_sql(""" SELECT rental_date, title, first_name, last_name
                FROM rental
                INNER JOIN inventory
                ON inventory.inventory_id = rental.inventory_id
                INNER JOIN film
                ON film.film_id = inventory.film_id
                INNER JOIN customer
                ON rental.customer_id = customer.customer_id
                ORDER BY title """, 
            connection).head(55)
```
